```python
#q18

toss_decisionsseason = matches.groupby(['season', 'toss_decision'])['toss_decision'].count().unstack()
toss_decisionsseason.plot(kind='bar', figsize=(11, 6))
plt.xlabel("Season")
plt.ylabel("Number of Toss Decisions")
plt.title("Toss Decision Across Seasons")
plt.xticks(rotation=45)
plt.legend(title="Toss Decision")
plt.show()
```

```python
#q19
#do it again using line graph
dismissal_count=deliveries['dismissal_kind'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(dismissal_count.index, dismissal_count.values,color='pink')
plt.xlabel("Dismissal Kind")
plt.ylabel("Frequency")
plt.title("Distribution of Dismissal Kind")
plt.xticks(rotation=90)
plt.show()
```

```python
#q20
topscorer=deliveries.groupby('batsman')['total_runs'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.bar(topscorer.index, topscorer.values,color='brown')
plt.xlabel("batsman")
plt.ylabel("runs")
plt.title("Top 10 IPL runners")
plt.xticks(rotation=45)
plt.show()
```

```python
#q21
MOM=matches.groupby('player_of_match')['player_of_match'].count().head(10)
plt.bar(MOM.index, MOM.values,color='grey')
plt.xlabel("Player")
plt.ylabel("Number of Player of the Match Awards")
plt.title("Top 10 Men of the Match")
plt.xticks(rotation=45)
plt.show()
```

```python
#q22
totalteam=pd.concat([matches['team1'],matches['team2']])
totalteamcnt=totalteam.value_counts()
print(totalteamcnt)
```

```python
#q23
total_matches = pd.concat([matches['team1'], matches['team2']]).value_counts()
winning_matches = matches['winner'].value_counts()
winning_rate = (winning_matches / total_matches) * 100
plott= pd.DataFrame({'Total Matches': total_matches,
                         'Winning Matches': winning_matches,
                         'Winning Rate': winning_rate})
plott.figure(figsize=(11,6))
plott.bar()
```

```python
#q24
winteam = matches['winner'].value_counts()
plt.figure(figsize=(11, 6))
plt.bar(winteam.index, winteam.values, color='green')
plt.xlabel("Teams")
plt.ylabel("Number of Matches Won")
plt.title("Distribution of Winning Teams'")
plt.xticks(rotation=90)
plt.show()
```

```python
#q25
totmatch=pd.concat([matches['team1'],matches['team2']]).value_counts()
winmatch=matches['winner'].value_counts()
RATIO=(totmatch)/(winmatch)
print(RATIO)
```

```python
#q26
tosswin = matches[matches['toss_winner'] == matches['winner']]

tossdecision = tosswin.groupby(['winner', 'toss_decision'])['toss_decision'].count().unstack()

tossdecision.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.xlabel("Teams")
plt.ylabel("Number of Toss Decisions")
plt.title("Toss Decision Choice of Teams")
plt.xticks(rotation=45)
plt.legend(title="Toss Decision")
plt.show()
```
